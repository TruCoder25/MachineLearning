# Linear Regression

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ddf = fetch_california_housing()
```

```python
dataset = pd.DataFrame(ddf.data)
```

```python
dataset.columns = ddf.feature_names
```

```python
dataset.head()
```

```python
##Independent Features(X) and dependent Features(y)

X = dataset
y = ddf.target
```

```python
y
```

```python
X
```

```python
## Traintest Split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
#implementing the LinearRegression model
#first Standardizing the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train) #we do here the fit_transform
```

```python
X_test = scaler.transform(X_test) ## here not because we need to standardize the whole data so that the whole data is in same scale,, example for the first cookie the the reciepe is given for the second cookie there is no reciepe,,, same reciepe is formed
```

```python
from sklearn.linear_model import LinearRegression
##cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse = cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
## prediction

pred = regression.predict(X_test)
```

```python
import seaborn as sns
sns.displot(pred-y_test,kind='kde')
pred
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,pred)
```
